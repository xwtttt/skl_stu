```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
faces = fetch_lfw_people(min_faces_per_person=60)
```

```python
faces.images.shape
```

```python
X = faces.data
```

```python
fig, axes = plt.subplots(2,3
                        ,figsize=(10,8)                        
                        ,subplot_kw = {"xticks":[],"yticks":[]} #不要显示坐标轴
)
# for i, ax in enumerate(axes.flat):
#     ax.imshow(faces.images[i, :, :], cmap="gray")
for i in range(3):
    axes[0, i].imshow(faces.images[i, :, :], cmap="gray")
    axes[1, i].imshow(faces.images[i, :, :], cmap="gray")
```

```python
pca = PCA(150)      #设置模型参数
```

```python
V = pca.fit(X)      #拟合模型
```

```python
X_reshape = V.transform(X)      #应用模型
```

```python
X_return = pca.inverse_transform(X_reshape)     #转回
```

```python
fig, axes = plt.subplots(3, 8, figsize=(15,8), subplot_kw={"xticks":[],"yticks":[]})

for i in range(8):    
    axes[0, i].imshow(faces.images[i,:,:] ,cmap="gray")
    axes[1, i].imshow(X_return[i, :].reshape(62,47), cmap="gray")
    axes[2, i].imshow(V.components_[i, :].reshape(62,47),cmap="gray")
```
